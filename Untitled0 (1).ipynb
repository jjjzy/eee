```
import math

def condEntropy(x, y):
  uniqueY = list(np.unique(y))
    
  result = 0

  for item in uniqueY:
    temp = []
    for i in range(len(y)):
      if y[i] == item:
        temp.append(i)
    
    newList = []
    for each in temp:
      newList.append(x[each])

    temp2 = np.unique(newList)  
    entropy = 0
    for items in temp2:
      pi = (float)((newList).count(items)/len(newList))
      entropy -= pi * math.log(pi, 2)

    pi = (float)((y).count(item)/len(y))
    result +=  pi * entropy
  return result

entropyy = []
for i in range(9):
  entropyy.append(condEntropy(list(train.iloc[:,i]), list(train.iloc[:,9])))

print(entropyy)
```

```
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

parameters = {'max_depth':range(1,30)}

clf = GridSearchCV(estimator = tree.DecisionTreeClassifier(),
                        param_grid = parameters,
                        scoring = 'accuracy', 
                        cv = 5, 
                        verbose = 0,
                        n_jobs = -1)

clf.fit(X=train.iloc[:, : 9], y=train.iloc[:,9])

print (clf.best_score_, clf.best_params_) 

print(clf.cv_results_['mean_test_score'])
print(clf.cv_results_['params'])

best_estim = clf.best_estimator_

best_estim.predict(test.iloc[:, : 9])
accuracy_score(test.iloc[:,9], best_estim.predict(test.iloc[:, : 9]))
```

```
from sklearn.ensemble import RandomForestClassifier

parameters = {'max_depth':range(1,20),
              'n_estimators':range(1,20)
              }

clf = GridSearchCV(estimator = RandomForestClassifier(),
                        param_grid = parameters,
                        scoring = 'accuracy', 
                        cv = 5, 
                        verbose = 0,
                        n_jobs = -1)


clf.fit(X=train.iloc[:, : 9], y=train.iloc[:,9])

print(clf.best_score_, clf.best_params_) 

print(clf.cv_results_['mean_test_score'])
print(clf.cv_results_['params'])

best_estim = clf.best_estimator_

best_estim.predict(test.iloc[:, : 9])
accuracy_score(test.iloc[:,9], best_estim.predict(test.iloc[:, : 9]))
```
